```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression

# Function to calculate the squared error cost for linear regression
def squared_error_cost_linear(m, b, x, y):
    n = len(x)
    cost = 0.0
    for i in range(n):
        y_pred = m * x[i] + b  # Prediction using the current slope (m) and intercept (b)
        cost += (y_pred - y[i]) ** 2  # Squared error
    cost /= (2 * n)  # Average cost
    return cost

# Function to calculate the squared error cost for logistic regression
# Note: This is not the actual cost function for logistic regression
def squared_error_cost_logistic(w, b, x, y):
    m = len(x)
    cost = 0.0
    for i in range(m):
        z = w * x[i] + b  # Linear combination
        y_pred = 1 / (1 + np.exp(-z))  # Sigmoid function to get probability
        cost += (y_pred - y[i]) ** 2  # Squared error
    cost /= (2 * m)  # Average cost
    return cost

# Function to plot the squared error cost function for linear regression
def plot_squared_error_cost_linear(x_train, y_train):
    m_range = np.linspace(-10, 10, 100)  # Range of slope values
    b_range = np.linspace(-10, 10, 100)  # Range of intercept values
    M, B = np.meshgrid(m_range, b_range)  # Create a grid of slope and intercept values
    J = np.zeros_like(M)  # Initialize cost matrix

    # Calculate the cost for each pair of (m, b)
    for i in range(len(m_range)):
        for j in range(len(b_range)):
            J[i, j] = squared_error_cost_linear(M[i, j], B[i, j], x_train, y_train)

    # Plotting the cost function surface
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(M, B, J, cmap='viridis')
    ax.set_xlabel('m')  # Label for slope axis
    ax.set_ylabel('b')  # Label for intercept axis
    ax.set_zlabel('Cost')  # Label for cost axis
    ax.set_title('Squared Error Cost Function for Linear Regression')

# Function to plot the squared error cost function for logistic regression
def plot_squared_error_cost_logistic(x_train, y_train):
    w_range = np.linspace(-10, 10, 100)  # Range of weight values
    b_range = np.linspace(-10, 10, 100)  # Range of intercept values
    W, B = np.meshgrid(w_range, b_range)  # Create a grid of weight and intercept values
    J = np.zeros_like(W)  # Initialize cost matrix

    # Calculate the cost for each pair of (w, b)
    for i in range(len(w_range)):
        for j in range(len(b_range)):
            J[i, j] = squared_error_cost_logistic(W[i, j], B[i, j], x_train, y_train)

    # Plotting the cost function surface
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(W, B, J, cmap='viridis')
    ax.set_xlabel('w')  # Label for weight axis
    ax.set_ylabel('b')  # Label for intercept axis
    ax.set_zlabel('Cost')  # Label for cost axis
    ax.set_title('Squared Error Cost Function for Logistic Regression')

# Sample data
x_train = np.array([0, 1, 2, 3, 4, 5], dtype=np.longdouble).reshape(-1, 1)  # Feature data
y_train = np.array([0, 0, 0, 1, 1, 1], dtype=np.longdouble)  # Target data

# Linear Regression using scikit-learn
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)  # Train the linear regression model

# Logistic Regression using scikit-learn
logistic_reg = LogisticRegression()
logistic_reg.fit(x_train, y_train)  # Train the logistic regression model

# Plotting the squared error cost function for linear regression
plot_squared_error_cost_linear(x_train, y_train)

# Plotting the linear regression prediction line
plt.plot(x_train, linear_reg.predict(x_train), color='blue', label='Linear Regression')
plt.legend()

# Plotting the squared error cost function for logistic regression
plot_squared_error_cost_logistic(x_train, y_train)

# Plotting the logistic regression prediction probabilities
plt.plot(x_train, logistic_reg.predict_proba(x_train)[:, 1], color='green', label='Logistic Regression')
plt.legend()

# Display the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
x = np.array(x0 + x1).reshape(-1, 1)
y = np.array(y0 + y1)

# Create an instance of the logistic regression model
model = LogisticRegression()

# Fit the model to the data
model.fit(x, y)

# Generate the curve points for plotting
x_curve = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_curve = model.predict_proba(x_curve)[:, 1]

# Plot the data points
plt.plot(x0, y0, marker='o', linestyle='', label='cant vote')
plt.plot(x1, y1, marker='x', linestyle='', label='can vote')

# Plot the logistic regression curve
plt.plot(x_curve, y_curve, label='logistic regression curve')

# Set plot labels and title
plt.ylabel('Eligibile to vote?')
plt.xlabel('Age')
plt.title('Logistic Regression Example')



# Find the x value corresponding to y=0.5(threshold)
x_threshold = x_curve[np.where(np.abs(y_curve - 0.5) == np.min(np.abs(y_curve - 0.5)))][0]

# Draw the vertical line
plt.axvline(x=x_threshold, color='black', linestyle='--')

# Fill the regions on each side of the threshold line
plt.fill_betweenx([0, 1], x_threshold[0], max(x_curve), color='green' ,alpha = 0.3, label='can vote')
plt.fill_betweenx([0, 1], min(x_curve), x_threshold[0], color='red', alpha=0.3, label='cant vote')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
X = df.drop('target' , axis = 1)
y = df['target']
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create an instance of the logistic regression model
logreg = LogisticRegression()

# Fit the model on the training data
logreg.fit(X_train, y_train)

# Predict the target variable for the test data
y_pred = logreg.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model is:", round(accuracy * 100, 2), "%")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)  #(df - df.min()) / (df.max() - df.min())
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

X_train = X_train.T
Y_train = y_train.reshape(1, X_train.shape[1])

X_test = X_test.T
Y_test = y_test.reshape(1, X_test.shape[1])
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def model(X, Y, learning_rate, iterations):
    m = X_train.shape[1]
    n = X_train.shape[0]

    W = np.random.randn(n, 1)
    B = 0

    cost_list = []

    for i in range(iterations):
        Z = np.dot(W.T, X) + B
        A = sigmoid(Z)

        cost = -(1/m) * np.sum(Y*np.log(A) + (1-Y)*np.log(1-A))

        # Gradient Descent
        dW = (1/m) * np.dot(A-Y, X.T)
        dB = (1/m) * np.sum(A - Y)

        W = W - learning_rate * dW.T
        B = B - learning_rate * dB

        # Keeping track of our cost function value
        cost_list.append(cost)

        if i % (iterations/10) == 0:
            print("cost after", i, "iteration is:", cost)

    return W, B, cost_list

def accuracy(X, Y, W, B):

    Z = np.dot(W.T, X) + B
    A = sigmoid(Z)

    A = A > 0.5

    A = np.array(A, dtype = 'int64')

    acc = (1 - np.sum(np.absolute(A - Y))/Y.shape[1])*100

    print("Accuracy of the model is : ", round(acc, 2), "%")
```

```python
iterations = 20000
learning_rate = 0.01
W, B, cost_list = model(X_train, Y_train, learning_rate = learning_rate, iterations = iterations)
```
